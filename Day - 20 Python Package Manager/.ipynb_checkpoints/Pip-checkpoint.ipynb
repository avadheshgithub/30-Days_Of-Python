# Pip - Python package manager

### What is PIP?

PIP stands for Preferred Installer Program. It is used to install different Python packages. A package in Python is a module or a collection of modules that can be imported and used in your application. Instead of writing every utility from scratch, you can install and use existing packages to save time and effort.

### 1. Installing PIP

```python
pip install pip
```

```python
pip --version
```

```python
pip list
```

```python
 # To generate a list of installed Python packages with their versions, use:
    pip freeze
```

### 2. Uninstalling Packages

```python
pip uninstall packagename
```

### 3. Reading from a URL

```python
pip install requests
```

```python
import requests

# Reading a text file from a website
url = 'https://www.w3.org/TR/PNG/iso_8859-1.txt'
response = requests.get(url)
print(response.text)

# Reading JSON data from an API
url = 'https://restcountries.com/v2/all'
response = requests.get(url)
countries = response.json()
print(countries[:1])
```
